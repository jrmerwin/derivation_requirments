```
from pathlib import Path
import os, sys, json
import pandas as pd
import numpy as np

# Adjust if needed.
HERE = Path.cwd()
BASE = HERE if (HERE / 'helpers').exists() else HERE.parent
HELPERS = BASE / 'helpers'
DATA = BASE / 'data'
OUT = BASE / 'handoff_outputs'
OUT.mkdir(exist_ok=True)

# If unpacked in the original project root, use ./src. Otherwise fall back to src_snapshot/src.
SRC = BASE / 'src'
if not SRC.exists() and (BASE / 'src_snapshot' / 'src').exists():
    SRC = BASE / 'src_snapshot' / 'src'

for p in [BASE, SRC, HELPERS]:
    if p.exists() and str(p) not in sys.path:
        sys.path.insert(0, str(p))

print('BASE:', BASE)
print('SRC:', SRC, 'exists=', SRC.exists())
print('HELPERS:', HELPERS, 'exists=', HELPERS.exists())
print('DATA:', DATA, 'exists=', DATA.exists())
print('OUT:', OUT)
```

```
helper_order = [
    'DEU_GR_Experiment_01H_depth_metric_repair.py',
    'DEU_GR_Experiment_01L_fixed_epoch_anchor_experiment.py',
    'DEU_GR_Experiment_01M_lapse_useful_budget.py',
    'DEU_GR_Experiment_01N_radial_lapse_profile.py',
    'DEU_GR_Experiment_01O_radial_lapse_residual.py',
    'DEU_GR_Experiment_01P_layer3_holonomy_extraction.py',
    'DEU_GR_Experiment_01N_topological_disclination_holonomy.py',
    'DEU_GR_Experiment_01Q_coherent_topology_stitch_layer3.py',
    'DEU_GR_Experiment_01R_anchor_locked_stitch_snap_audit.py',
    'DEU_GR_Experiment_01S_exterior_boundary_goldilocks_audit.py',
]

for name in helper_order:
    path = HELPERS / name
    print('loading', path)
    if not path.exists():
        print('  MISSING')
        continue
    exec(path.read_text(encoding='utf-8'), globals())
print('done')
```
